# Project report

The solution code of the exercise can be found in the following files:
* Solution.ipynb: this Jupyter Notebook contains the DDPG function to train and test the model using the Unity environment.
* ddpg_agent.py: this code file contains the Agent class with all the functionality for the agent to act and learn in the environment. The code also contains the ReplayBuffer class with all the code for implementing Experience Replay.
* model.py: contains the Python code for the architecture of the neural network that is used to train the agent. The Actor class contains the network for training the Actor Policy. The Critic class contains the network for training the Critic Policy.

## Learning Algorithm

I used the DDPG algorithm and I started with the code from the DDPG-pendulum example. And applied batch normalisation on the inputs.
The network of the Actor consists of a hidden layer of size 400 with RELU and another hidden layer of size 300 with RELU.
The network of the Critic consists of a hidden layer of size 404 with RELU and another hidden layer of size 300 with RELU.
I used dropout layers for both the Actor and the Critic network.
I also used gradient clipping when training the critic network.

## Ideas for Future Work

I would take the following steps to further improve the model:
* Parameter tuning to further improve performance of the model. Training time was too long to try out all optimal combinations of hyper parameters.
* Try other algorithms like like PPO, A3C, and D4PG.
* Try different architectures for the actor and critic networks: deeper network, larger layers.
* Learn from the pixel data instead of receiving only state variables as input. The pixel data will contain much more information. So it would require a more complex learning architecture with convolutional layers, but ultimately result in a better performing agent.
